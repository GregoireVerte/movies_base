```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

movies_df = pd.read_csv("tmdb_movies.csv")
```

```python
q3_vote_count = movies_df["vote_count"].quantile(0.75)
```

```python
filtered_movies = movies_df[movies_df["vote_count"] > q3_vote_count]
```

```python
top_movies = filtered_movies.sort_values(by="vote_average", ascending=False)
```

```python
top_ten_movies = top_movies.head(10)
```

```python
print(top_ten_movies[["title", "vote_average", "vote_count","release_date"]])
```

```python
movies_df["release_year"] = pd.to_datetime(movies_df["release_date"], errors='coerce').dt.year
filtered_df = movies_df[(movies_df["release_year"] >= 2010) & (movies_df["release_year"] <= 2016)]
```

```python
grouped = filtered_df.groupby("release_year")[["revenue", "budget"]].mean()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(grouped.index, grouped["revenue"], color="blue", alpha=0.8, label="Revenues")
ax.plot(grouped.index, grouped["budget"], color="red", linewidth=2, label="Budget")

ax.set_ylim(0, grouped["revenue"].max() * 1.1)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{x/1e6:.0f}M"))

ax.set_xticks(grouped.index)
ax.set_xticklabels(grouped.index.astype(int))

ax.set_title("Średni przychód i budżet filmów w latach 2010-2016")
ax.set_xlabel("Rok")
ax.set_ylabel("Kwota (w mln $)")
ax.legend(loc="upper left", bbox_to_anchor=(1.03, 1))

plt.show()
```

```python
genres_df = pd.read_csv("tmdb_genres.csv", names=["genre_id", "genre_name"], header=0)
```

```python
merged_df = movies_df.merge(genres_df, on="genre_id", how="left")

print(merged_df[["title", "genre_id", "genre_name"]].head(15))
```

```python
genre_counts = merged_df["genre_name"].value_counts()
most_common_genre = genre_counts.idxmax()
most_common_count = genre_counts.max()

print(f"Najczęściej pojawiajacy się gatunek filmu: {most_common_genre}")
print(f"Liczba filmów w tym gatunku w bazie to: {most_common_count}")
```

```python
genre_avg_runtime = merged_df.groupby("genre_name")["runtime"].mean()

longest_runtime_genre = genre_avg_runtime.idxmax()
longest_runtime_value = genre_avg_runtime.max()

print(f"Gatunek z najdłuższym średnim czasem trwania to: {longest_runtime_genre}")
print(f"Najdłuższy średni czas trwania filmów tego gatunku: {longest_runtime_value:.1f} minut")
```

```python
from matplotlib.ticker import MaxNLocator

filtered_genre_df = merged_df[merged_df["genre_name"] == longest_runtime_genre]
runtime_counts = filtered_genre_df["runtime"].value_counts().sort_index()
max_count = runtime_counts.max()

plt.figure(figsize=(12, 8))
plt.hist(filtered_genre_df["runtime"].dropna(), bins=30, color='red', edgecolor='black')
plt.title(f'Histogram czasu trwania filmów z gatunku: {longest_runtime_genre}')
plt.xlabel('Czas trwania (minuty)')
plt.ylabel('Liczba filmów')
plt.yticks(range(0, max_count + 1, 1))

plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True, prune='lower', nbins=5))
plt.grid(True, which='both')
plt.show()
```
